### Plots dataset 16 points

```python
df_plot_16 = pd.read_csv('/content/result16.csv', sep = ";")
df_plot_16
```

```python
data_line = df_plot_16
f, axes = plt.subplots(nrows=3, ncols=2, figsize = (15, 15))
sns.lineplot(data=data_line, x='generation', y='fitness', hue='select', \
             estimator='mean', ci=99, err_style='band', legend=True, ax=axes[0, 0])
sns.lineplot(data=data_line, x='generation', y='fitness', hue='crossover', \
             estimator='mean', ci=99, err_style='band', legend=True, ax=axes[0, 1])
sns.lineplot(data=data_line, x='generation', y='fitness', hue='mutate', \
             estimator='mean', ci=99, err_style='band', legend=True,ax=axes[1, 0])
sns.lineplot(data=data_line, x='generation', y='fitness', hue='co_p', \
             estimator='mean', ci=99, err_style='band', legend=True, ax=axes[1, 1])
sns.lineplot(data=data_line, x='generation', y='fitness', hue='mu_p', \
             estimator='mean', ci=99, err_style='band', legend=True, ax=axes[2, 0])
sns.lineplot(data=data_line, x='generation', y='fitness', hue='elitism', \
             estimator='mean', ci=99, err_style='band', legend=True, ax=axes[2, 1])
plt.show()
```

```python
data_box = df_plot_16[df_plot_16.generation == 199]
f, axes = plt.subplots(6, figsize = (10, 12))
sns.boxplot(x="fitness", y="select", orient="h", data = data_box, ax=axes[0])
sns.boxplot(x="fitness", y="crossover", orient="h", data = data_box, ax=axes[1])
sns.boxplot(x="fitness", y="mutate", orient="h", data = data_box, ax=axes[2])
sns.boxplot(x="fitness", y="co_p", orient="h", data = data_box, ax=axes[3])
sns.boxplot(x="fitness", y="mu_p", orient="h", data = data_box, ax=axes[4])
sns.boxplot(x="fitness", y="elitism", orient="h", data = data_box, ax=axes[5])
plt.show() 
```

```python
data_box = df_plot_16[(df_plot_16.time < 0.4)]
f, axes = plt.subplots(6, figsize = (10, 12))
sns.boxplot(x="time", y="select", orient="h", data = data_box, ax=axes[0])
sns.boxplot(x="time", y="crossover", orient="h", data = data_box, ax=axes[1])
sns.boxplot(x="time", y="mutate", orient="h", data = data_box, ax=axes[2])
sns.boxplot(x="time", y="co_p", orient="h", data = data_box, ax=axes[3])
sns.boxplot(x="time", y="mu_p", orient="h", data = data_box, ax=axes[4])
sns.boxplot(x="time", y="elitism", orient="h", data = data_box, ax=axes[5])
plt.show() 
```
