### Libraries

```python
import zipfile
import urllib.request
import pandas as pd
pd.set_option('display.max_rows', 200)

import seaborn as sns
import matplotlib.pyplot as plt
```

### read files

```python
# download the file with 16 cities
url = 'https://github.com/anabeatrizfig/Comp_Intelligence_Optimization/blob/main/output/result16.zip?raw=true'
filename = 'result16.zip'
urllib.request.urlretrieve(url, filename)

# Unzip the downloaded file
zip_ref = zipfile.ZipFile("result16.zip")
zip_ref.extractall()
zip_ref.close()

# download the file with 29 cities
url = 'https://github.com/anabeatrizfig/Comp_Intelligence_Optimization/blob/main/output/result29.zip?raw=true'
filename = 'result29.zip'
urllib.request.urlretrieve(url, filename)

# Unzip the downloaded file
zip_ref = zipfile.ZipFile("result29.zip")
zip_ref.extractall()
zip_ref.close()
```

```python
# create the dataframes
df_plot_16 = pd.read_csv('result16.csv', sep = ";")
df_plot_29 = pd.read_csv('result29.csv', sep = ";")
df_plot_561 = pd.read_csv('https://raw.githubusercontent.com/anabeatrizfig/Comp_Intelligence_Optimization/main/output/result561_choices_1000generations.csv', sep = ",")
```

### Plots dataset 16 points

```python
# Figure 1 of the Report
data_box = df_plot_16[(df_plot_16.generation == 199) & (df_plot_16['elitism'] == True)]

f, axes = plt.subplots(5, figsize = (10, 12))
sns.boxplot(x="fitness", y="select", orient="h", data = data_box, ax=axes[0])
sns.boxplot(x="fitness", y="crossover", orient="h", data = data_box, ax=axes[1])
sns.boxplot(x="fitness", y="mutate", orient="h", data = data_box, ax=axes[2])
sns.boxplot(x="fitness", y="co_p", orient="h", data = data_box, ax=axes[3])
sns.boxplot(x="fitness", y="mu_p", orient="h", data = data_box, ax=axes[4])

plt.show() 
```

```python
# Figure 2 of the Report
data_line = df_plot_16.loc[df_plot_16['elitism'] == True]

f, axes = plt.subplots(nrows=3, ncols=2, figsize = (15, 15))
sns.lineplot(data=data_line, x='generation', y='fitness', hue='select', \
             estimator='mean', ci=99, err_style='band', legend=True, ax=axes[0, 0])
sns.lineplot(data=data_line, x='generation', y='fitness', hue='crossover', \
             estimator='mean', ci=99, err_style='band', legend=True, ax=axes[0, 1])
sns.lineplot(data=data_line, x='generation', y='fitness', hue='mutate', \
             estimator='mean', ci=99, err_style='band', legend=True,ax=axes[1, 0])
sns.lineplot(data=data_line, x='generation', y='fitness', hue='co_p', \
             estimator='mean', ci=99, err_style='band', legend=True, ax=axes[1, 1])
sns.lineplot(data=data_line, x='generation', y='fitness', hue='mu_p', \
             estimator='mean', ci=99, err_style='band', legend=True, ax=axes[2, 0])

plt.show()
```

```python
# Figure 3 of the report
data_line_elite = df_plot_16
data_box_elite = df_plot_16[(df_plot_16.generation == 199)]

f, axes = plt.subplots(2, figsize = (8, 8))
sns.lineplot(data=data_line_elite, x='generation', y='fitness', hue='elitism', \
             estimator='mean', ci=99, err_style='band', legend=True, ax=axes[0])
sns.boxplot(x="fitness", y="elitism", orient="h", data = data_box_elite, ax=axes[1])
plt.show()
```
